```
from zipfile import ZipFile
file_name = "/content/drive/My Drive/CovidDataProcessed.zip"
with ZipFile(file_name,'r') as zip:
  zip.extractall()
  print('Extraction completed!')
```

```
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D,MaxPooling2D,Flatten,Dense
```

```
## Creating the NN 
model=Sequential()
## Convolutional Layer 1
model.add(Conv2D(128,(3,3),padding='valid',input_shape=(64,64,3),activation='relu'))
## Pooling 1
model.add(MaxPooling2D(pool_size=(2,2)))
## Convolutional Layer 2
model.add(Conv2D(256,(3,3),padding='valid',activation='relu'))
## Pooling 2
model.add(MaxPooling2D(pool_size=(2,2)))
## Flattening
model.add(Flatten())
## Hidden Layer
model.add(Dense(512,activation='relu'))
## Output Layer
model.add(Dense(1,activation='sigmoid'))
```

```
## Compiling CNN
model.compile(optimizer='adam', loss= 'binary_crossentropy', metrics=['accuracy'])
```

```
## Preprocess the Images
from keras.preprocessing.image import ImageDataGenerator

train_datagen= ImageDataGenerator(rescale = 1./255,
                                  shear_range = 0.2,
                                  zoom_range = 0.2,
                                  horizontal_flip = True)

test_datagen= ImageDataGenerator(rescale= 1./255,
                                 shear_range = 0.2,
                                 zoom_range = 0.2,
                                 horizontal_flip = True)
```

```
training_set = train_datagen.flow_from_directory('/content/CovidData/Train',
                                                 target_size= (64,64),
                                                 batch_size=32,
                                                 class_mode= 'binary')
```

```
test_set = test_datagen.flow_from_directory('/content/CovidData/Test',
                                            target_size = (64,64),
                                            class_mode = 'binary',
                                            batch_size = 32)
```

```
## Training the model
model.fit_generator(training_set,
                    steps_per_epoch = 18,
                    epochs = 500,
                    validation_data = test_set,
                    validation_steps = 4)
```

```
## To save the model 
model.save("/content/drive/My Drive/CNN_COVID.model")
```
